### ORDER BY

```python
query = """
SELECT bike_number, birth_date, duration
FROM trips
ORDER BY duration 
LIMIT 10
"""
```

```python
db_interact(query)
```

#### How to order a subset

```python
query = """
SELECT bike_number, birth_date, duration
FROM trips
LIMIT 10
ORDER BY duration 
"""
```

```python
db_interact(query)
```

```python
query = """
SELECT * FROM (
SELECT bike_number, birth_date, duration
FROM trips
LIMIT 10
)
ORDER BY duration 
"""
```

```python
db_interact(query)
```

The sort order is ascending **ASC** by default to reverse the order: **DESC** for descending.

```python
query = """
SELECT * FROM (
SELECT bike_number, birth_date, duration
FROM trips
ORDER BY duration DESC
LIMIT 10
)
"""
```

```python
db_interact(query)
```

### Excercise 

Find the bike number and starting station id of the longest tip

```python
query = """
SELECT * FROM (
SELECT bike_number, start_station, duration
FROM trips
ORDER BY duration DESC
LIMIT 1
)
"""
```

```python
db_interact(query)
```

### PROJECTIONS with **AS**

Attributes can be renamed with AS.

```python
query = """
SELECT bike_number AS id, birth_date as bday
FROM trips
LIMIT 1
"""
```

```python
db_interact(query)
```
